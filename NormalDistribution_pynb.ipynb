```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(0)

# Generate 20 different means and standard deviations
means = np.random.uniform(-5, 5, 20)
std_devs = np.random.uniform(-3, 4.5, 20)

# Create a figure for each combination of mean and standard deviation
for i in range(20):
    mean = means[i]
    std_dev = std_devs[i]

    # Print the mean and standard deviation
    print(f'Distribution {i+1} - Mean: {mean:.2f}, Standard Deviation: {std_dev:.2f}')

    # Generate a range of values for the x-axis
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
    y = norm.pdf(x, mean, std_dev)

    # Plot the normal distribution
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'Normal Distribution\nmean = {mean:.2f}, std dev = {std_dev:.2f}')
    plt.fill_between(x, y, alpha=0.2)

    # Add titles and labels
    plt.title(f'Normal Distribution Curve - Distribution {i+1}')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Set random seed for reproducibility
np.random.seed(0)
means = np.random.uniform(0, 10, 20)
std_devs = np.random.uniform(1, 3, 20)

x_min = min(means) - 4 * max(std_devs)
x_max = max(means) + 4 * max(std_devs)
x = np.linspace(x_min, x_max, 1000)

y_max = 1 / (min(std_devs) * np.sqrt(2 * np.pi))

for i in range(20):
    mean = means[i]
    std_dev = std_devs[i]


    print(f'Distribution {i+1} - Mean: {mean:.2f}, Standard Deviation: {std_dev:.2f}')

    y = norm.pdf(x, mean, std_dev)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=f'Normal Distribution\nmean = {mean:.2f}, std dev = {std_dev:.2f}')
    plt.fill_between(x, y, alpha=0.2)

    plt.title(f'Normal Distribution Curve - Distribution {i+1}')
    plt.xlabel('Value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)

    plt.xlim(x_min, x_max)
    plt.ylim(0, y_max)

    plt.show()
```
