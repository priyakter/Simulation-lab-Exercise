```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data set
data = [1.2, 2.5, 2.8, 3.3, 3.7, 4.1, 5.6, 6.8, 7.9, 8.5,3.4,7.1,4.3,2.8,3.1,7.8,8.3,5.6,3.7,1.9]

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)


# Generate a range of values for the x-axis
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = norm.pdf(x, mean, std_dev)

# Plot the normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Normal Distribution\nmean = {mean:.2f}, std dev = {std_dev:.2f}')
plt.fill_between(x, y, alpha=0.2)

# Mark mean and standard deviations
plt.axvline(mean, color='r', linestyle='--', label='Mean (μ)')
plt.axvline(mean - std_dev, color='g', linestyle='--', label='1 Std Dev (μ - σ)')
plt.axvline(mean + std_dev, color='g', linestyle='--')
plt.axvline(mean - 2*std_dev, color='b', linestyle='--', label='2 Std Dev (μ - 2σ)')
plt.axvline(mean + 2*std_dev, color='b', linestyle='--')
plt.axvline(mean - 3*std_dev, color='y', linestyle='--', label='3 Std Dev (μ - 3σ)')
plt.axvline(mean + 3*std_dev, color='y', linestyle='--')

# Add titles and labels
plt.title('Normal Distribution Curve')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
